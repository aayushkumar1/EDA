```python
#here i'm finding the count of each room type in the entire NYC.
plt.rcParams['figure.figsize'] = (8, 5)

ax = sns.countplot(y="room_type", hue='neighbourhood_group', data=dt, palette='bright')

total = len(dt['room_type'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_width() / total)
    x = p.get_width() + 0.02
    y = p.get_y() + p.get_height() / 2
    ax.annotate(percentage, (x, y), ha='left', va='center')

plt.title('Count of Each Room Type in Entire NYC')
plt.xlabel('Room Counts')
plt.ylabel('Room Types')
plt.xticks(rotation=90)
plt.show()
```


The chosen count plot effectively visualizes room type distribution across NYC neighborhoods. Insights include understanding prevalent room types in different areas. This understanding aids in strategic decisions regarding property investment, marketing, and pricing, potentially enhancing customer satisfaction and business performance. However, over-reliance on certain room types or imbalanced distribution may lead to missed opportunities and reduced flexibility, necessitating a balanced approach to sustain long-term growth.

```python
#let's check the price relation with room types.
title = 'Price per Room Type for Properties under $175'
data_filtered = dt.loc[dt['price'] < 175]
f, ax = plt.subplots(figsize=(8, 6))
sns.boxplot(x='room_type', y='price', data=data_filtered, notch=True, showmeans=True,
            meanprops={"marker":"s","markerfacecolor":"white","markeredgecolor":"black"})
plt.title(title)
plt.show()

title = 'Price per Room Type for Properties over $175'
data_filtered = dt.loc[dt['price'] > 175]
f, ax = plt.subplots(figsize=(8, 6))
sns.boxplot(x='room_type', y='price', data=data_filtered, notch=True, showmeans=True,
            meanprops={"marker":"s","markerfacecolor":"white","markeredgecolor":"black"})
plt.title(title)
plt.show()
```

The box plots effectively visualize the relationship between price and room types for properties priced under $175 and over $175, respectively. Insights include understanding the distribution of prices across different room types and identifying potential variations in pricing based on room type. These insights can inform pricing strategies, marketing efforts, and property positioning to maximize revenue and customer satisfaction.






```python
#let's see the price relation with number of reviews per month.
sns.set_palette('muted')

x = 'reviews_per_month'
y = 'price'

title = 'Price Relation to Number of Reviews per Month for Properties under $175'
data_filtered = dt.loc[(dt['price'] < 175) & (dt['reviews_per_month'] < 30)]
f, ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(x=x, y=y, data=data_filtered)
plt.title(title)
plt.show()

sns.set_palette('muted')

x = 'reviews_per_month'
y = 'price'

title = 'Price Relation to Number of Reviews per Month for Properties over $175'
data_filtered = dt.loc[(dt['price'] > 175) & (dt['reviews_per_month'] < 30)]
f, ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(x=x, y=y, data=data_filtered)
plt.title(title)
plt.show()
```

The scatter plots visualize the relationship between price and number of reviews per month for properties priced under $175 and over $175, respectively. Insights include understanding how price correlates with review frequency and identifying potential trends or patterns in customer feedback and pricing. These insights can help optimize pricing strategies and identify opportunities for improving customer satisfaction and property performance.
