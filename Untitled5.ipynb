```python
# we drop some unnecessary column such as id,last_review
dt=dt.drop(['id','last_review'],axis=1)
```

```python
#filling the null values
dt.fillna({'name':'not present','host_name':'absent','reviews_per_month':0},inplace=True)
```

```python
#let's see the short summary of the data again with to check is there any rebundancy left.
dt.describe()
```

```python
#let's check the columns where price==0 and availability_365==0
wrong_listings=dt[(dt['price']==0) & (dt['availability_365']==0)]
wrong_listings
```

I can see two wrong values in price and availability_365. which is 0,so it'll affect the business.the reason for this could be the host has set the wrong price or maybe the host has disconnected his listing.

```python
#let's exclude property with listed price of 0
dt=dt.loc[dt['price']>0]
```

```python
#let's check again the summary of the data
dt.describe()
```

```python

# Set the seaborn color palette
sns.set_palette("muted")

# Create a figure and axis object
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Plot each distribution on its own subplot
sns.histplot(dt['price'], ax=axes[0, 0])
axes[0, 0].set_title('Price Distribution')

sns.histplot(dt['minimum_nights'], ax=axes[0, 1])
axes[0, 1].set_title('Minimum Nights Distribution')

sns.histplot(dt['reviews_per_month'], ax=axes[0, 2])
axes[0, 2].set_title('Reviews per Month Distribution')

sns.histplot(dt['calculated_host_listings_count'], ax=axes[1, 0])
axes[1, 0].set_title('Calculated Host Listings Count Distribution')

sns.histplot(dt['number_of_reviews'], ax=axes[1, 1])
axes[1, 1].set_title('Number of Reviews Distribution')

sns.histplot(dt['availability_365'], ax=axes[1, 2])
axes[1, 2].set_title('Availability 365 Distribution')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
